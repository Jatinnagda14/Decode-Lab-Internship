```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('C:\\Users\\LENOVO\\OneDrive\\Documents\\Decode Lab Internship\\employee_data_dirty_1500.csv')
```

```python
df.info(),df.head(10), df.describe()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df.isnull().sum()
```

```python
print(df["EmployeeID"].duplicated().sum())
df=df.drop_duplicates(subset=["EmployeeID"], keep="first")
print(df["EmployeeID"].duplicated().sum())
df.shape
```

```python
df['Phone']=df['Phone'].astype(str)
df['EmployeeID']=df['EmployeeID'].astype(str)
df['Age']=df['Age'].astype(int)
```

```python

df['JoiningDate']=pd.to_datetime(df['JoiningDate'], errors='coerce',format='mixed')
df['JoiningDate'].isna().sum()
```

```python
(df["Phone"].str.len() != 10).sum()
```

```python
df.loc[df["Phone"].str.len() != 10, "Phone"] = "Invalid Number"
df["Name"] = df["Name"].str.strip().str.title()
```

```python
invalid_email = ~df["Email"].str.contains(
    r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$",
    regex=True
)

df.loc[invalid_email, "Email"] = "Invalid Email"
print((df["Email"] == "Invalid Email").sum())
```
